```python
df = pd.read_csv('file2.csv')
df.head()
```

```python
df['DISCOVERY YEAR'] = pd.DatetimeIndex(df['DISCOVERY_DATE_Greg']).year
```

```python
df = df.drop(columns=['Unnamed: 0','STAT_CAUSE_CODE','DISCOVERY_DATE_Greg','CONT_DATE_Greg','DISCOVERY_DATE','CONT_DATE','DISCOVERY YEAR'])
df
```

```python
df = df.dropna()
```

```python
dls = TabularDataLoaders.from_csv('file2.csv', y_names="STAT_CAUSE_DESCR",
    cat_names = ['STATE','DISCOVERY_MONTH','DISCOVERY_DAY_OF_WEEK'],
    cont_names = ['FIRE_YEAR', 'LATITUDE', 'LONGITUDE','FIRE_SIZE','DISCOVERY_TIME','CONT_TIME','DAYS_SPENT','HRS_SPENT'],
    procs = [Categorify, FillMissing, Normalize])
```

```python
splits = RandomSplitter(valid_pct=0.2)(range_of(df))
```

```python
to = TabularPandas(df, procs=[Categorify, FillMissing,Normalize],
                   cat_names = ['STATE','DISCOVERY_MONTH','DISCOVERY_DAY_OF_WEEK'],
                   cont_names = ['FIRE_YEAR', 'LATITUDE', 'LONGITUDE','FIRE_SIZE','DISCOVERY_TIME','CONT_TIME','DAYS_SPENT','HRS_SPENT'],
                   y_names='STAT_CAUSE_DESCR',
                   y_block=CategoryBlock,
                   splits=splits)
```

```python
to.xs.iloc[:2]
```

```python
dls = to.dataloaders(bs=64)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
```

```python
learn.fit_one_cycle(50)
```

```python
learn.show_results()
```

```python
row, clas, probs = learn.predict(df.iloc[0])
```
